```python
base = pl.read_csv('atacadao_*.csv', infer_schema_length=10000)


base_agrupada = base.with_columns([
    pl.col.data.str.strptime(dtype=pl.Date, format='%Y-%m-%d %H:%M:%S%.f')
]).with_columns([
    pl.col.data.dt.strftime('%d/%b').alias('dia_mes')
]).group_by(['fabricante','nome_prod', 'dia_mes']).agg([
    pl.col.preço.median()
])

base_dia = base_agrupada.sort('dia_mes').pivot(values = 'preço', index = ['fabricante','nome_prod'], columns = 'dia_mes')
base_dia
```

```python
base_desconto = base_dia.with_columns([
    pl.concat_list([x for x in base_dia.columns if x.endswith('May')]).alias('preços')
]).select([
    pl.col.fabricante,
    pl.col.nome_prod,
    pl.col.preços.list.max().alias('alto'),
    pl.col.preços.list.min().alias('baixo'),
    pl.col.preços.list.median().alias('normal'),
]).with_columns([
    (1-(pl.col.baixo/pl.col.alto)).alias(r'% desconto'),
    (pl.col.alto-pl.col.baixo).alias('R$ desconto'),
]).sort('% desconto', descending=True)

base_desconto
```

```python
base_estoque = base_dia.with_columns([
    pl.concat_list([x for x in base_dia.columns if x.endswith('May')]).list.len().alias('dias_lidos'),
    (pl.concat_list([x for x in base_dia.columns if x.endswith('May')]).list.drop_nulls().list.len()).alias('dias_estoque')
]).select([
    'fabricante',
    'nome_prod',
    'dias_lidos',
    (pl.col.dias_lidos-pl.col.dias_estoque).alias('dias_falta')
]).group_by(['fabricante']).agg([
    pl.col.nome_prod.count().alias('# SKUs'),
    pl.col.dias_falta.sum().alias('falta_total'),
    (pl.col.dias_falta.sum()/pl.col.nome_prod.count()).alias('média falta'),
]).sort('# SKUs', descending=True)

base_estoque.to_pandas()
```
